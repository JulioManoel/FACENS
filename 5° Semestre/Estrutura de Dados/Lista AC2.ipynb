## Exercício 1 

Escreva uma função chamada **compara** que receba como parâmetro uma pilha e uma fila. Ela deve retornar True caso as duas estruturas passadas como parâmetro sejam iguais e False caso contrário. Um pilha e uma fila são consideradas iguais tanto quando os elementos estão na mesma ordem ou em ordem contrária. Por exemplo, a pilha A=[0,1,2,3,4] é igual as filas B = [0,1,2,3,4] e C = [4,3,2,1,0].

 - Você deve usar as classes Pilha e Fila ensinadas na disciplina e disponibilizadas juntamente com este notebook. 

- As únicas operações permitidas para acessar os valores da pilha e fila são, respectivamente, as operações pop e dequeue. 

- Você deve garantir que ao final da função, a pilha e fila passadas como parâmetro estejam intactas, isto é, como os mesmos valores passados como entrada. 

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################
class Node:
  """
  Nó de uma pilha e fila dinamica
  """
    
  def __init__(self, data = None, next = None):
    self.data = data
    self.next = next

class Pilha:
  """
  Pilha dinamica
  """
    
  def __init__(self):
    self.head = None
    self.tamanho = 0

  def push(self, data):
    """
    Adiciona um elemento no topo da pilha
    """
        
    # Se a pilha esta vazia, o element head ira receber o novo noh
    if self.head is None:
      self.head = Node(data)

    # Se não cria um novo noh com o dado informado e sera usado como novo head  
    else:
      temp = Node(data, self.head)
      self.head = temp
        
    self.tamanho += 1

  def pop(self):
    """
    Retorna e remove o elemento do topo da pilha
    """

    # Verifica se o head é vazio    
    if self.head is None:
      print("Pilha vazia")
      return None

    # Exclui o ultimo noh        
    data = self.head.data
    self.head = self.head.next
        
    self.tamanho -= 1
    return data
    
  def size(self):
    """
    Retorna a quantidade de elementos da pilha
    """
    return self.tamanho

  def isEmpty(self):
    """
    Retorna True se a fila esta vazia
    """
    return self.head is None
            
  def __str__(self):
    """
    Retorna uma string com as informacoes da pilha
    quando o objeto e chamado dentro de um print() 
    ou dentro de um str(). 
    """
    # Cria uma pilha auxiliar
    auxPilha = Pilha()
        
    # Retira elementos da pilha preenchendo a auxiliar em ordem reversa
    while not self.isEmpty():
      auxPilha.push( self.pop() )            
            
    # Cria string com elementos da auxiliar enquanto preenche a pilha original
    strPilha = '['
    while not auxPilha.isEmpty():
      
      top = auxPilha.pop()
        
      # Preenche a pilha original
      self.push(top)
            
      # Guarda o top em uma string
      strPilha += ' ' + str(top)
            
    strPilha += ' ]'
    return strPilha

class Fila:
  """
  Fila dinamica
  """

  def __init__(self, data  = None):
    self.front = None
    self.rear = None
    self.size = 0
        
    # Se dado for passado, usa a funcao enqueue para adicionar esse dado
    if data is not None:
      self.enqueue(data)

  def enqueue(self, data):
    """
    Adiciona um elemento no final da fila
    """

    # Cria um novo noh com o dado informado
    tempNode = Node(data)
            
    # Se a fila esta vazia, o elemento rear receberá o noh e front receberá rear
    if self.rear is None:
      self.rear = tempNode
      self.front = self.rear

    # Se não proximo no de rear receberá o noh e rear receberá o noh        
    else:
      self.rear.next = tempNode
      self.rear = tempNode
        
    self.size += 1

  def dequeue(self):
    """
    Retorna e remove o elemento no inicio da fila
    """
        
    # Verifica se aa fila esta vazia
    if self.size == 0:
      print("Fila vazia")
      return None
        
    # Pega dado do elemento front e front receberá o elemento seguinte de front
    data = self.front.data
    self.front = self.front.next
        
    # Se o novo elemento front for None (ou seja, a fila ficou vazia)
    # O elemento rear tambem deve ser setado como none
    if self.front is None:
      self.rear = None
        
    self.size -= 1
    return data
    
  def getSize(self):
    """
    Retorna a quantidade de elementos da fila
    """
    return self.size

  def isEmpty(self):
    """
    Retorna True se a fila esta vazia
    """
    return self.front is None

  def getFront(self):
    """
    Retorna o primeiro elemento da lista sem remover
    """
        
    if self.front is None:
      print("Pilha vazia")
      return None

    return self.front.data

  def clear(self):
    """
    Limpa a fila
    """
    while not self.isEmpty():
      data = self.dequeue()
            
  def __str__(self):
    """
    Retorna uma string com as informacoes da fila
    quando o objeto e chamado dentro de um print() 
    ou dentro de um str(). 
    """
    
    # Cria uma string com os elementos da fila
    strfila = '['

    # Retira o elemento da fila, coloca na string e volta para a fila
    for i in range(self.size):
      front = self.dequeue()
      strfila += ' ' + str(front)
      self.enqueue(front)
    
    strfila += ' ]'
    return strfila


def compara(pilha, fila):
  """
  Verifica se a pilha e fila são iguais em ordem ou em ordem contraria
  """

  # Verifica se as duas filas tem o mesmo tamanho
  if pilha.size() != fila.getSize():
    return False

  # Cria uma pilhaAux e uma variavel que verifica se os elementos são diferentes
  pilhaAux = Pilha()
  diferent = False

  # Armazena o valor do pop e dequeue e verifica se são diferentes
  while not pilha.isEmpty():
    pilhaNum = pilha.pop()
    filaNum = fila.dequeue()

    # Armazena os valor da fila na propria fila e da pilha na pilhaAux
    fila.enqueue(filaNum)
    pilhaAux.push(pilhaNum)

    if pilhaNum != filaNum:
      diferent = True
      break
  
  # Se for diferente verifica termina de passar todos os elementos da pilha
  if diferent:
    diferent = False
    while not pilha.isEmpty():
      fila.enqueue(fila.dequeue())
      pilhaAux.push(pilha.pop())

    # Realiza o mesmo procedimento do while mais acima
    while not pilhaAux.isEmpty():
      pilhaNum = pilhaAux.pop()
      filaNum = fila.dequeue()
      fila.enqueue(filaNum)
      pilha.push(pilhaNum)

      if pilhaNum != filaNum:
        diferent = True
        break

  # Organiza a pilha
  while not pilhaAux.isEmpty():
    pilha.push(pilhaAux.pop())
    fila.enqueue(fila.dequeue())

  return False if diferent else True 

##########################################################################

# testando a função criada
pilha1 = Pilha()
pilha1.push(0)
pilha1.push(1)
pilha1.push(2)
pilha1.push(3)
pilha1.push(4)

# testando a função criada
pilha2 = Pilha()
pilha2.push(4)
pilha2.push(3)
pilha2.push(2)
pilha2.push(1)
pilha2.push(0)

fila1 = Fila()
fila1.enqueue(0)
fila1.enqueue(1)
fila1.enqueue(2)
fila1.enqueue(3)
fila1.enqueue(4)

fila2 = Fila()
fila2.enqueue(4)
fila2.enqueue(3)
fila2.enqueue(2)
fila2.enqueue(1)
fila2.enqueue(0)

fila3 = Fila()
fila3.enqueue(4)
fila3.enqueue(3)
fila3.enqueue(2)
fila3.enqueue(0)
fila3.enqueue(1)

comp1 = compara(pilha1, fila1)
comp2 = compara(pilha1, fila2)
comp3 = compara(pilha2, fila1)
comp4 = compara(pilha2, fila2)
comp5 = compara(pilha2, fila3)

print('Resultado retornado:')
print('\tComparacao 1?: ', comp1)
print('\tComparacao 2?: ', comp2)
print('\tComparacao 3?: ', comp3)
print('\tComparacao 4?: ', comp4)
print('\tComparacao 5?: ', comp5)

print('\n' + 20*'-')
print('Resultado esperado:')
print('\tComparacao 1?: True')
print('\tComparacao 2?: True')
print('\tComparacao 3?: True')
print('\tComparacao 4?: True')
print('\tComparacao 5?: False')
````

Faça uma classe chamada **ListaNSE**. Essa classe deve implementar uma **lista não sequencial e estática**. Essa lista deve ter o mesmo comportamento de uma **lista não sequencial e dinâmica**, mas deve ser guardada dentro de uma lista em Python com capacidade limitada.

Essa classe deve possui os seguintes métodos:

- inserir(x, i): insere x na posição i da estrutura
- remover(i): remove o item que está na posição i da estrutura e retorna seu valor 
- buscar(x): buscar o item x na lista e retorna sua posição
- size(): retorna o número de elementos da lista
- clear(): remove os elementos da lista
- \_\_str\_\_(): imprime a estrutura

- Você deve usar a lista do Python apenas para inicializar a estrutura. Depois disso, nenhum método do Python pode ser usado, como append, insert, etc.

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################
class ListaNSE:
  """
  Lista não Sequencial Estática
  """
  
  def __init__(self, capacite):
    self.tamanho = 0
    self.capacite = capacite
    self.lista = [None] * capacite

  def insert(self, value, index):
    if index < 0 or index >= self.capacite:
      print("Index fora do Range")
      return None
      
    elif self.tamanho == self.capacite:
      print("Lista Cheia")
      return
    
    if self.lista[index] == None:
      self.lista[index] = value
    else:
      for i in range(self.capacite-1, index, -1):
        if self.lista[i] == None:
          self.lista[i] = self.lista[i-1];
          self.lista[i-1] = None
      self.lista[index] = value

    self.tamanho += 1

  def remove(self, index):
    if index < 0 or index >= self.capacite:
      print("Index fora do Range")
      return
    elif self.tamanho <= 0:
      print("Lista Vazia")
      return
    
    value = self.lista[index]
    self.lista[index] = None
    self.tamanho -= 1
    return value

  def size(self):
    return self.tamanho

  def clear(self):
    self.tamanho = 0
    self.lista = [None] * self.capacite
    

  def __str__(self):
    strLista = '[ '

    for i in range(self.capacite):
      if self.lista[i] != None:
        strLista += str(self.lista[i]) + ' '

    strLista += ']'
    return strLista

##########################################################################

# testa a Lista
lista = ListaNSE(5)        

# inserir no inicio da lista
lista.insert(5,0)
print( 'insert(5,0): ', lista )

lista.insert(10,0)
print( 'insert(10,0): ', lista )

lista.insert(24,2)
print( 'insert(24,2): ', lista )

lista.insert(25,1)
print( 'insert(25,1): ', lista )

lista.remove(2)
print( 'remove(2): ', lista )

print( 'Size: ', lista.size())

lista.clear()
print( 'Clear: ', lista )

print('\n' + 20*'-')
print('Resultado esperado:')
print( 'insertHead(5,0): [5]')
print( 'insertHead(10,0): [10,5]')
print( 'insert(24,2): [10,5,24]')
print( 'insert(25,1): [10,25,5,24]')
print( 'remove(2): [10,25,24]')
print( 'Size: 3')
print( 'Clear: []')
````

Modifique a lista duplamente ligada ensinada nessa disciplina e fornecida juntamente com este trabalho, alterando seu nome para **ListaInsertion**.

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################
class Node:
  """
  Noh de uma lista duplamente ligada
  """
  def __init__(self, data = None, next = None, prev = None):
    self.data = data
    self.next = next # apontador para o noh da frente
    self.prev = prev # apontador para o noh de tras

class ListaInsertion:
  """
  Lista duplamente ligada
  """

  def __init__(self, data  = None):
    self.head = None
    self.tail = None
    self.size = 0

  def insertHead(self, data):
    """
    Adiciona um elemento no inicio da lista
    """
        
    # cria um novo noh com o dado informado
    tempNode = Node(data)

    # se a lista esta vazia, o element head
    # ira receber o novo noh.
    # Nesse caso,o elemento tail devera apontar para 
    # o mesmo noh do elemento head        
    if self.size==0:
      self.head = tempNode
      self.tail = tempNode
          
    else:
        
      # o proximo elemento na lista a frente do novo noh sera o noh head antigo
      tempNode.next = self.head
      
      # o elemento atras do noh head antigo sera o novo noh
      self.head.prev = tempNode
      
      # o novo noh head sera o novo noh
      self.head = tempNode
        
    # incrementa a quantidade de elementos da lista
    self.size += 1
        
  def insertTail(self, data):
      """
      Adiciona um elemento no final da lista
      """

      # cria um novo noh com o dado informado
      tempNode = Node(data)
          
      # se a lista esta vazia, o element tail
      # ira receber o novo noh.
      # Nesse caso,o elemento head devera apontar para 
      # o mesmo noh do elemento tail
      if self.size==0:
        self.tail = tempNode
        self.head = self.tail
          
      else:
          
        # o tail antigo ficara atras do novo noh e, por isso,
        # o apontador prev do novo noh deve apontar para o tail antigo
        tempNode.prev = self.tail
        
        # o noh tail sera substituido pelo novo noh e, por isso,
        # seu apontador next deve apontar para o novo noh
        self.tail.next = tempNode
        
        # o novo noh tail sera o novo noh
        self.tail = tempNode
      
      self.size += 1
        
  def busca(self, node, value):
    """
    Realiza uma busca binaria, retornando a posição onde posição para inserção
    """
    primeiro = 1
    ultimo = self.size
    i = primeiro

    while True:
      meio = (primeiro + ultimo) // 2

      # Loop para possicionar o noh na posição correta
      while i < meio:
        node = node.next
        i += 1
        
      while i > meio:
        node = node.prev
        i -= 1

      media = self.getMediaPalavra(node.data)

      # Verifica se é possivel a inserção nessa posição
      if (node.prev is None or value >= media) and (node.next is None or value <= self.getMediaPalavra(node.next.data)):
        return meio

      elif value < media:
        ultimo = meio - 1

      else:
        primeiro = meio + 1     
        
  def insert(self, data):
    """
    Adiciona um elemento na posição correta da lista
    """       
    
    # Obtem a média da frase a ser inserida
    num = self.getMediaPalavra(data)

    # Cria um novo noh com a frase informada
    tempNode = Node(data)
           
    # Se lista esta vazia, o tail receberá o noh e head deverá apontar para tail
    if self.size == 0:
      self.tail = tempNode
      self.head = self.tail

    # Se não, será tratado a frase
    else:

      # Realiza a busca para indentificar onde pode ser inserido
      prevNode = self.head
      position = self.busca(prevNode, num)

      # Coloca o prevNode no lugar correto para a inserção
      id = 1
      while id < position:
        prevNode = prevNode.next
        id += 1

      # Verifica se a inserção é para o lado esquerdo ou direito e a realiza
      if num >= self.getMediaPalavra(prevNode.data):
        tempNode.prev = prevNode
        if prevNode.next is None:
          self.tail = tempNode
        else:
          tempNode.next = prevNode.next
          prevNode.next.prev = tempNode
        prevNode.next = tempNode
        
      else:
        tempNode.next = prevNode
        if prevNode.prev is None:
          self.head = tempNode
        else:
          tempNode.prev = prevNode.prev
          prevNode.prev.next = tempNode
        prevNode.prev = tempNode

    self.size += 1

  def getMediaPalavra(self, frase):
    """
    Retornar a media de caracteres por palavra da frase passada
    """

    # Divide a frase em palavras
    palavras = frase.split()

    # Calcula incrementa media com o tamanho das palavras
    media = 0
    for palavra in palavras:
      media += len(palavra)

    # Retorna a media redondado para baixo
    return int(media/len(palavras))

  def repeticoes(self, value, primeiro = 1, ultimo = None, i = 1, node = None):
    """
    Recebe um número e usa uma busca binária recursiva modificada,
    para que retorne a quantidade de strings que contém a mesma quantidade média de caracteres.
    """
    if ultimo is None:
      ultimo = self.size

    if node is None:
      node = self.head
    
    meio = (primeiro + ultimo) // 2

    # Loop para possicionar o noh na posição correta
    while i < meio:
      node = node.next
      i += 1
        
    while i > meio:
      node = node.prev
      i -= 1

    # Verifica se os valores são iguais
    if value == self.getMediaPalavra(node.data):
      rep = 1

      while not node.prev is None and self.getMediaPalavra(node.prev.data) == value:
        node = node.prev

      while not node.next is None and self.getMediaPalavra(node.next.data) == value:
        node = node.next
        rep += 1
      return rep

    # Caso não possua retorna -1 
    elif (node.next is None or value < self.getMediaPalavra(node.next.data)) and (node.prev is None or value > self.getMediaPalavra(node.prev.data)):
      return -1

    # Caso não seja uma posição possivel de inserção realiza a busca recursiva
    elif value < self.getMediaPalavra(node.data):
      return self.repeticoes(value, primeiro, meio-1, i, node)
    else:
      return self.repeticoes(value, meio+1, ultimo, i, node)

  def subLista(self, a, b):
    """
    Esse método deve receber dois parâmetros a e b e retornar uma nova lista apenas com os valores que estão entre as posições a e b.
    """
    # Cria uma subLista
    subLista = ListaInsertion()
    i = 1
    node = self.head

    if a < 1:
      print('Valor A tem que ser maior que 1')
      return subLista
    
    if self.size == 0:
      print('Lista Vazia')
      return subLista
    
    if b > self.size:
      print('Valor B incorreto maior que o tamanho da lista')
      return subLista

    while i < a:
      node = node.next
      i += 1

    while i <= b:
      subLista.insert(node.data)
      node = node.next
      i += 1

    # Retorna a subLista
    return subLista

  def compara(self, positionA, positionB):
    """
    verifica se duas strings são iguais. Duas strings são iguais se tiverem a mesma quantidade média de caracteres.
    """

    # Verifica se a lista esta vazia
    if self.size == 0:
      print('Lista Vazia')
      return False

    # Verifica se A e B estão no intervalo da lista    
    if positionB < 1 or positionB > self.size or positionA < 1 or positionA > self.size:
      print('Valor A e B tem que estar entre 1 e tamanho maximo da lista')
      return False

    # Verifica qual é o melhor camanho para chegar até o valor
    if positionA < int(self.size/2):
      i = self.size
      node = self.tail
      while i > positionA: 
        node = node.prev
        i -= 1
    
    else:
      i = 1
      node = self.head
      while i < positionA: 
        node = node.next
        i += 1

    # Armazena o valor da positionA
    data1 = node.data
    
    # Verifica qual é a melhor forma de chegar até o valor
    if positionB < int(self.size/2):
      i = self.size
      node = self.tail
      while i > positionB: 
        node = node.prev
        i -= 1
    
    else:
      i = 1
      node = self.head
      while i < positionB: 
        node = node.next
        i += 1

    # Armazena valor da positionB
    data2 = node.data

    # Retornar se são iguais ou não
    return True if self.getMediaPalavra(data1) == self.getMediaPalavra(data2) else False

  def getSize(self):
    """
    Retorna a quantidade de elementos da lista
    """
    return self.size

  def isEmpty(self):
    """
    Retorna True se a lista esta vazia
    """
    return self.head is None
    
  def clear(self):
    """
    Limpa a lista
    """ 
    while not self.isEmpty():
      self.remove(0)
            
  def __str__(self):
    """
    Retorna uma string com as informacoes da lista
    quando o objeto e chamado dentro de um print() 
    ou dentro de um str().    
    A lista nao possui restricao de acesso aos seus elementos. 
    Por isso, ela nao precisa ser destruida para que seus elementos sejam 
    visitados como na fila e na pilha
    """

    strfila = '['
       
    if self.size > 0:
            
      # comecao pelo head
      tempNode  = self.head
      valor = tempNode.data
        
      # guarda o valor do head em uma string
      strfila += str(valor)
            
      # percorre o resto da lista ate que o apontador next do noh seja nulo
      # o que indica que chegou no tail
      while tempNode.next is not None:
                
        # vai para o proximo noh
        tempNode = tempNode.next
                
        valor = tempNode.data 
        
        # guarda o valor em uma string
        strfila += ', ' + str(valor)
            
      strfila += ']'
        
      return strfila

##########################################################################

# testa o exercicio
lista = ListaInsertion()        

# inserir no inicio da lista
lista.insert("Centro Universitario Facens")
lista.insert("Eu fui para um shopping em Sorocaba")
lista.insert("A menina gosta de jogar")
lista.insert("O menino gosta de andar")
lista.insert("Tá tá tá")

print("\nLista:\n", lista)

qtd1 = lista.repeticoes(3)
qtd2 = lista.repeticoes(2)
qtd3 = lista.repeticoes(30)
print("Qtd. repeticoes 1: ", qtd1)
print("Qtd. repeticoes 2: ", qtd2)
print("Qtd. repeticoes 3: ", qtd3)

print('\n' + 20*'-')
print('Resultado esperado:')
print("\nLista:\n")
print('[Tá tá tá, A menina gosta de jogar, ', end="")
print('O menino gosta de andar, ', end="")
print('Eu fui para um shopping em Sorocaba, ', end="")
print('Centro Universitario Facens]', end="")
````

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################
class Node:
  def __init__(self, data, next=None):
    self.data = data
    self.next = next

class pilhaFila:
  """
  Pilha juntamente com Fila aproventando a estrutura do deque (fila dupla) dinâmico
  """

  def __init__(self):
    self.primeiro = None
    self.ultimo = None
    self.tamanho = 0
  
  def push(self,valor):
    """
    Insere um elemento no final da fila
    """
    noh = Node(valor)

    if self.tamanho == 0:
      self.primeiro = noh
      self.ultimo = noh
    else:
      self.ultimo.next = noh
      self.ultimo = noh

    # Aumenta a qtd. de elementos
    self.tamanho += 1

  def enqueue(self, valor):
    """
    Insere um elemento no final da fila
    """
    self.push(valor)

  def dequeue(self):
    """
    Exclui o primeiro elemento
    """
    if self.isEmpty():
      print('pilhaFila vazia')
      return None
    
    nohAExcluir = self.primeiro

    # Armazena o valor
    valor = nohAExcluir.data
    
    # Atualiza quem está na frente
    self.primeiro = nohAExcluir.next
    
    # Diminui o numero de elementos
    self.tamanho -= 1

    # Deleta o noh da memoria
    del nohAExcluir
    
    # Retorna o valor
    return valor

  def pop(self):
    """
    Exclui o ultimo elemento
    """
    if self.isEmpty():
      print('pilhaFila vazia')
      return None
    
    nohAExcluir = self.ultimo
    
    # Armazena o valor
    valor = nohAExcluir.data

    # Se a pilhaFila tiver tamanho 1, ao excluir o noh, ele ficara vazio
    if self.tamanho == 1:
      self.primeiro = None
      self.ultimo = None
    
    # Se não, atualiza quem esta atras. Precisa tambem atualizar o next do no que sera o novo rear
    else:
      nohAux = self.primeiro
      novoRear = None
      
      # Percorre a pilhaFila ate achar o elemento anterior ao rear.
      while nohAux.next is not None:
        novoRear = nohAux
        nohAux = nohAux.next

      # Troca o rear
      self.ultimo = novoRear

    # Diminui o numero de elementos
    self.tamanho -= 1
      
    # Deleta o noh da memoria
    del nohAExcluir

    # Retorna valor
    return valor

  def front(self):
    """
    Retorna o primeiro elemento
    """
    if self.primeiro is None:
      print("Pilha vazia")
      return None

    # Retornado o dado do primeiro noh
    return self.primeiro.data

  def rear(self):
    """
    Retorna o ultimo elemento
    """
    if self.ultimo is None:
      print("Pilha vazia")
      return None
    
    # Retornado o dado do ultimo noh
    return self.ultimo.data

  def top(self):
    """
    Retorna o ultimo elemento
    """
    return self.rear()

  def size(self):
    """
    Retorna o tamanho
    """
    return self.tamanho

  def isEmpty(self):
    """
    Verifica se não possui elementos
    """
    if self.tamanho == 0:
      return True
    else:
      return False

  def clear(self):
    """
    Apaga dos os elementos
    """
    while not self.isEmpty():
      self.dequeue()
  
  def __str__(self):
    
    # pilhaFila auxiliar para imprimir a fila original
    dequeAux = pilhaFila()
    
    # String que ira guardar os valores
    info = '['

    while not self.isEmpty():
      valor = self.dequeue()
      
      # Insere na pilhaFIla auxiliar
      dequeAux.push( valor )

      info = info + ' ' + str(valor)

    info = info + ' ]'

    # Laço para recuperar a fila
    while not dequeAux.isEmpty():
      valor = dequeAux.dequeue()
      self.push(valor)
    return info

##########################################################################
       
# testando a PilhaFila
pilhaFila1 = pilhaFila()

pilhaFila1.push(5)
print(pilhaFila1)

pilhaFila1.enqueue(3)
print(pilhaFila1)

pilhaFila1.push(9)
print(pilhaFila1)

print( 'Size: ', pilhaFila1.size() )

valor = pilhaFila1.pop()
print(pilhaFila1)

valor = pilhaFila1.dequeue()
print(pilhaFila1)

print( 'Vazia?: ', pilhaFila1.isEmpty() )

valor = pilhaFila1.pop()
print(pilhaFila1)

print( 'Vazia?: ', pilhaFila1.isEmpty() )

valor = pilhaFila1.pop()
print(pilhaFila1)

pilhaFila1.push(2)
print(pilhaFila1)

pilhaFila1.push(14)
print(pilhaFila1)

print(pilhaFila1.front())
print(pilhaFila1.rear())
print(pilhaFila1.top())

print( 'Vazia?: ', pilhaFila1.isEmpty() )

pilhaFila1.clear()
print( 'Vazia?: ', pilhaFila1.isEmpty() )

print('\n' + 20*'-')
print('Resultado esperado:')
print("\nLista:\n")
print('[ 5 ]')
print('[ 3, 5 ]')
print('[ 3, 5, 9 ]')
print('Size:  3')
print('[ 3 5 ]')
print('[ 3 ]')
print('Vazia?:  False')
print('[ ]')
print('Vazia?:  True')
print('pilhaFila1 vazia')
print('[ ]')
print('[ 2 ]')
print('[ 2 14 ]')
print('2')
print('14')
print('14')
print('Vazia?:  False')
print('Vazia?:  True')
````
